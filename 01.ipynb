```python
with open('input2022_01.csv', 'r') as elve:
    elves = elve.read()
```

```python
highest_elve_calories = 0
elves = elves.rstrip('\n')
elve_list = elves.split('\n\n')
for calories in elve_list:
    calories_list = calories.split('\n')
    elve_calories = sum(list(map(int, calories_list)))
    if elve_calories > highest_elve_calories:
        highest_elve_calories = elve_calories
print(highest_elve_calories)

```
